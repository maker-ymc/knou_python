```python
# 반지름, 높이 값 할당
rad=int(input("반지름 값을 입력하세요: "))
hei=int(input("높이 값을 입력하세요: "))

if rad > 0 and hei>0 :

  # 부피&겉넓이 계산
  vol= 1/3*3.14*rad**2*hei
  suf = 3.14*rad**2+3.14*rad*hei

  #결과 출력
  print("원뿔의 부피는", vol, ", 겉넓이는 ", suf, " 입니다")

else :
  print("반지름과 높이의 값을 모두 양수로 입력해주세요")
```

```python
# A, B, C 사용자 입력
A=int(input("A 입력 : "))
B=int(input("B 입력 : "))
C=int(input("C 입력 : "))

# A, B, C 중 가장 큰 수 출력 (내장함수 Max의 기능을 직접 만든다)
if A > B :
  if A>C:
    print(A)
  else :
    print(C)
else :
  if B>C :
    print(B)
  else :
    print(C)
```
